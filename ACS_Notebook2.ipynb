```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(123)

# Define die outcomes and probabilities
dice, probabilities, throws, outcomes, success = [1, 2, 3, 4, 5, 6], [1/6, 1/6, 1/6, 1/6, 1/6, 1/6], 2, [], 0

# Number of simulations
num_simulations = 100

# Lists to store outcomes for each simulation
simulation_outcomes = []

for _ in range(num_simulations):
    # Use np.random.choice to throw the die once and record the outcome
    outcomes = np.random.choice(dice, size=throws, p=probabilities)
    simulation_outcomes.append(outcomes)

# Calculate the probability of getting doubles in each simulation
doubles_probability = [1 if outcome[0] == outcome[1] else 0 for outcome in simulation_outcomes]

# Convert the count to probability by dividing by the total number of simulations
doubles_probability = np.cumsum(doubles_probability) / np.arange(1, num_simulations + 1)

# Set Seaborn style
sns.set(style="whitegrid")

# Plot the probability of getting doubles per simulation
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(1, num_simulations + 1), y=doubles_probability)
plt.xlabel("Number of Simulations")
plt.ylabel("Probability of Doubles")
plt.title("Probability of Getting Doubles in Two Dice Throws (Per Simulation)")
plt.grid(True)
plt.show()
```
